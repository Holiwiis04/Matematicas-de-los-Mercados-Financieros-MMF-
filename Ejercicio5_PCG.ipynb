```python
import numpy as np
import matplotlib.pyplot as plt

def wiener_process(n_steps, dt):
    """
    Generates a Wiener process (Brownian motion) with n_steps time steps and time step dt.
    """
    increments = np.random.normal(0, np.sqrt(dt), n_steps)
    wiener = np.cumsum(increments)
    return wiener

def ito_integral(mu, wiener, dt):
    """
    Calculates the Itô integral of mu * W(t) with respect to the Wiener process.
    """
    integral = np.zeros_like(wiener)
    for i in range(1, len(wiener)):
        integral[i] = integral[i-1] + mu * wiener[i-1] * (wiener[i] - wiener[i-1])
    return integral

# Parameters
mu = 0.1  # Expected return rate
T = 1.0   # Total time
n_steps = 1000  # Number of time steps
dt = T / n_steps  # Time step size

# Generate Wiener process
wiener = wiener_process(n_steps, dt)

# Calculate Itô integral
integral = ito_integral(mu, wiener, dt)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(np.linspace(0, T, n_steps), integral, label='Itô Integral')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Simulation of Itô Integral')
plt.legend()
plt.grid(True)
plt.show()
```
